```python
df_main['last_4overs_wickets'] = df_main.groupby(['match_id','inning'])['wicket'].apply(lambda x: x.rolling(24).sum())
```

```python
df_main['last_4overs_runs'] = df_main.groupby(['match_id','inning'])['runs'].apply(lambda x: x.rolling(24).sum())
```
